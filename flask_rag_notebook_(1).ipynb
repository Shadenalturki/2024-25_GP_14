```python
# -----------------------------
# Utility Functions
# -----------------------------
def save_chat_history(session_id: str, chat_history: ChatMessageHistory):
    """
    Save chat history to a JSON file named chat_history_<session_id>.json
    """
    filename = f"{CHAT_HISTORY_PREFIX}{session_id}.json"
    messages_to_save = []
    for msg in chat_history.messages:
        messages_to_save.append({"role": msg.type, "content": msg.content})
    try:
        with open(filename, "w", encoding="utf-8") as f:
            json.dump(messages_to_save, f)
        logging.debug(f"Saved chat history for session {session_id} to {filename}.")
    except Exception as e:
        logging.error(f"Error saving chat history for session {session_id}: {e}")


def load_chat_history(session_id: str) -> ChatMessageHistory:
    """
    Load chat history from chat_history_<session_id>.json if it exists.
    Returns a ChatMessageHistory object.
    """
    filename = f"{CHAT_HISTORY_PREFIX}{session_id}.json"
    chat_history = ChatMessageHistory()
    if os.path.isfile(filename):
        try:
            with open(filename, "r", encoding="utf-8") as f:
                messages = json.load(f)
            for m in messages:
                role = m["role"].lower()
                content = m["content"]
                if role in ["human", "user"]:
                    chat_history.add_user_message(content)
                elif role in ["assistant", "ai"]:
                    chat_history.add_ai_message(content)
                else:
                    chat_history.add_message(role, content)
            logging.debug(f"Chat history loaded for session {session_id}.")
        except Exception as e:
            logging.error(f"Error loading chat history for session {session_id}: {e}")
    return chat_history


def extract_text_from_pdf(file_content):
    try:
        with tempfile.NamedTemporaryFile(delete=False, suffix=".pdf") as pdf_file:
            pdf_file.write(file_content)
            pdf_file_path = pdf_file.name
        logging.debug(f"Temporary PDF file created at {pdf_file_path}")

        pdfreader = PdfReader(pdf_file_path)
        text = ''
        for page in pdfreader.pages:
            content = page.extract_text()
            if content:
                text += content

        os.unlink(pdf_file_path)
        logging.debug("Temporary PDF file removed")
        return text
    except Exception as e:
        logging.error("Error extracting text from PDF", exc_info=True)
        traceback.print_exc()
        raise e
```

```python
# -----------------------------
# Chain Definitions: Summarization, Translation, and Quiz Generation
# -----------------------------
summarize_prompt = PromptTemplate(
    input_variables=["context"],
    template="""
       You are a summarizer. Provide a final summary of the entire text as bullet points.
       Add a Generic Motivational Title.
       Summary must:
       - Cover all essential contents
       - Be comprehensive enough to replace the slides
       - Simplify definitions but keep accuracy
       - Include enumerations/examples in structured order

       Content: {context}

       Your Answer:
       Title:
       (Bullet points start here)
    """
)
summarize_chain = LLMChain(llm=llm, prompt=summarize_prompt)

translate_prompt = PromptTemplate.from_template(
    """
    Translate this text into Arabic, return only the Arabic Text:
    {text}
    Translation:
    """
)

RESPONSE_JSON = {
    "1": {"mcq": "", "options": {"a": "", "b": "", "c": ""}, "correct": ""},
    "2": {"mcq": "", "options": {"a": "", "b": "", "c": ""}, "correct": ""},
    "3": {"mcq": "", "options": {"a": "", "b": "", "c": ""}, "correct": ""}
}

quiz_prompt_template = PromptTemplate(
    input_variables=["text", "number", "response_json"],
    template=(
        """
Text: {text}

You are an expert quiz maker. Based on the above text, create {number} multiple-choice questions in JSON format.
Use the EXACT structure of keys as follows:
 {response_json}
        """.strip()
    )
)
quiz_chain = LLMChain(llm=llm, prompt=quiz_prompt_template)

quiz_pdf_prompt_template = PromptTemplate(
    input_variables=["text"],
    template=(
        """REQUIREMENTS:
- Generate EXACTLY 15 multiple-choice questions
- Each question MUST follow this format:
  Question X: [Question text]
  a) [Option 1]
  b) [Option 2]
  c) [Option 3]
  Correct Answer: [letter]

Text:
{text}

CRITICAL: Ensure ALL 15 QUESTIONS are generated. No fewer, no more.
        """
    )
)
quiz_pdf_chain = LLMChain(llm=llm, prompt=quiz_pdf_prompt_template)

def determine_number_of_questions(text):
    word_count = len(text.split())
    num_questions = min(max(1, word_count // 100), 15)
    logging.debug(f"Determined {num_questions} questions from {word_count} words.")
    return num_questions

def validate_and_format_json_output(raw_output):
    try:
        lines = raw_output.strip().split("\n\n")
        quiz_list = []
        for block in lines:
            block = block.strip()
            if block.startswith("{") and block.endswith("}"):
                try:
                    quiz_item = json.loads(block)
                    quiz_list.append(quiz_item)
                except json.JSONDecodeError:
                    logging.warning(f"Skipping invalid JSON block:\n{block}")
            else:
                logging.info(f"Ignoring non-JSON block:\n{block}")
        return {"quiz": quiz_list} if quiz_list else None
    except Exception as e:
        logging.error("Error in validate_and_format_json_output", exc_info=True)
        return None
```

```python
# -----------------------------
# Conversational RAG Chain Creation
# -----------------------------
def create_conversational_chain(api_key, documents, session_id):
    """
    Creates a new RAG chain for the given session_id,
    storing embeddings in db_<session_id>.
    """
    llm_chat = ChatGroq(groq_api_key=api_key, model_name="Gemma2-9b-It")

    text_splitter = RecursiveCharacterTextSplitter(chunk_size=5000, chunk_overlap=500)
    splits = text_splitter.split_documents(documents)

    vectorstore = Chroma.from_documents(
        documents=splits,
        embedding=embeddings,
        persist_directory=f"{DB_PREFIX}{session_id}"
    )
    retriever = vectorstore.as_retriever()

    from langchain.chains import create_history_aware_retriever
    contextualize_q_prompt = ChatPromptTemplate.from_messages([
        ("system", "Given chat history, rephrase the question if necessary."),
        MessagesPlaceholder("chat_history"),
        ("human", "{input}"),
    ])
    history_aware_retriever = create_history_aware_retriever(llm_chat, retriever, contextualize_q_prompt)

    qa_prompt = PromptTemplate(
        input_variables=["context", "input"],
        template=("""You are an assistant. Use the retrieved context to answer the user's query:

Context:
{context}

User Question: {input}

Answer:""")
    )
    question_answer_chain = create_stuff_documents_chain(llm_chat, qa_prompt)
    rag_chain = create_retrieval_chain(history_aware_retriever, question_answer_chain)

    chat_history = ChatMessageHistory()

    conversational_rag_chain = RunnableWithMessageHistory(
        rag_chain,
        lambda session: chat_history,
        input_messages_key="input",
        history_messages_key="chat_history",
        output_messages_key="answer"
    )

    return conversational_rag_chain, chat_history


def create_conversational_chain_from_chroma(api_key, session_id):
    """
    If a db_<session_id> folder already exists, create the chain from it.
    """
    llm_chat = ChatGroq(groq_api_key=api_key, model_name="Gemma2-9b-It")

    vectorstore = Chroma(
        persist_directory=f"{DB_PREFIX}{session_id}",
        embedding_function=embeddings
    )
    retriever = vectorstore.as_retriever()

    from langchain.chains import create_history_aware_retriever
    contextualize_q_prompt = ChatPromptTemplate.from_messages([
        ("system", "Given chat history, rephrase the question if necessary."),
        MessagesPlaceholder("chat_history"),
        ("human", "{input}"),
    ])
    history_aware_retriever = create_history_aware_retriever(llm_chat, retriever, contextualize_q_prompt)

    qa_prompt = PromptTemplate(
        input_variables=["context", "input"],
        template=("""You are an assistant. Use the retrieved context to answer the user's query:

Context:
{context}

User Question: {input}

Answer:""")
    )
    question_answer_chain = create_stuff_documents_chain(llm_chat, qa_prompt)
    rag_chain = create_retrieval_chain(history_aware_retriever, question_answer_chain)

    chat_history = ChatMessageHistory()

    conversational_rag_chain = RunnableWithMessageHistory(
        rag_chain,
        lambda session: chat_history,
        input_messages_key="input",
        history_messages_key="chat_history",
        output_messages_key="answer"
    )

    return conversational_rag_chain, chat_history
```

```python
# -----------------------------
# Flask Routes
# -----------------------------
# The following endpoints cover summarization, translation, quiz generation, document upload, and chatting.

# @app.route('/upload', methods=['POST'])
# def upload_pdf():
#     # Code for PDF upload without indexing (if needed)
#     pass

@app.route('/summarize', methods=['POST'])
def summarize():
    data = request.json
    if not data or 'text' not in data:
        return jsonify({"error": "No text provided"}), 400
    summary = summarize_chain.run({"context": data['text']})
    return jsonify({"summary": summary})


@app.route('/translate', methods=['POST'])
def translate():
    data = request.json
    if not data or 'text' not in data:
        return jsonify({"error": "No text provided"}), 400
    try:
        translation_llm = Together(model="meta-llama/Llama-3.3-70B-Instruct-Turbo-Free", temperature=0)
        translation_chain = LLMChain(llm=translation_llm, prompt=translate_prompt)
        translated_summary = translation_chain.run({"text": data['text']})
        return jsonify({"translated_summary": translated_summary})
    except Exception as e:
        traceback.print_exc()
        return jsonify({"error": str(e)}), 500


@app.route('/generate_quiz', methods=['POST'])
def generate_quiz():
    data = request.json
    if not data or 'text' not in data:
        return jsonify({"error": "No text provided"}), 400
    text = data['text'].strip()
    if not text:
        return jsonify({"error": "Empty text provided"}), 400
    number_of_questions = determine_number_of_questions(text)
    raw_output = quiz_chain.run({
        "text": text,
        "number": number_of_questions,
        "response_json": json.dumps(RESPONSE_JSON)
    })
    logging.debug(f"Raw quiz chain output: {raw_output}")
    quiz_result = validate_and_format_json_output(raw_output)
    if not quiz_result:
        return jsonify({"error": "Quiz generation failed. Invalid or incomplete output."}), 500
    return jsonify(quiz_result)


@app.route('/generate_quiz_all', methods=['POST'])
def generate_quiz_all():
    try:
        if 'file' not in request.files:
            return jsonify({"error": "No file uploaded"}), 400
        file = request.files['file']
        if not file.filename.lower().endswith('.pdf'):
            return jsonify({"error": "Only PDF files are allowed"}), 400
        extracted_text = extract_text_from_pdf(file.read())
        if not extracted_text:
            return jsonify({"error": "No text generated from file"}), 400
        quiz_output = quiz_pdf_chain.run({"text": extracted_text})
        logging.debug(f"Quiz PDF chain output: {quiz_output}")

        quiz_list = []
        raw_questions = quiz_output.split("Question ")
        for idx, block in enumerate(raw_questions[1:], start=1):
            lines = [l.strip() for l in block.strip().split("\n") if l.strip()]
            if not lines:
                continue
            first_line = lines[0]
            colon_index = first_line.find(":")
            question_text = first_line[colon_index+1:].strip() if colon_index != -1 else first_line

            options = {}
            correct_answer = None
            for line in lines[1:]:
                low = line.lower()
                if low.startswith(("a)", "b)", "c)")):
                    label = low.split(')')[0].strip()
                    option_text = line[2:].strip()
                    options[label] = option_text
                if low.startswith("correct answer:"):
                    correct_answer = line.split(":", 1)[1].strip()

            if len(options) == 3 and correct_answer:
                quiz_list.append({
                    "question": question_text,
                    "options": options,
                    "correct": correct_answer
                })
            if len(quiz_list) == 15:
                break

        if len(quiz_list) != 15:
            return jsonify({"error": f"Generated {len(quiz_list)} questions instead of 15"}), 500

        return jsonify({"quiz": quiz_list})
    except Exception as e:
        traceback.print_exc()
        return jsonify({"error": str(e)}), 500


@app.route('/upload-index', methods=['POST'])
def upload_documents():
    api_key = "gsk_..."  # Replace with your actual LLM key
    if 'files' not in request.files:
        return jsonify({"error": "No files provided"}), 400
    uploaded_files = request.files.getlist("files")
    documents = []
    session_id = str(uuid4())
    for uploaded_file in uploaded_files:
        if not uploaded_file.filename.lower().endswith('.pdf'):
            return jsonify({"error": f"Invalid file format: {uploaded_file.filename}"}), 400
        with tempfile.NamedTemporaryFile(delete=False, suffix=".pdf") as tmp:
            tmp.write(uploaded_file.read())
            tmp_path = tmp.name
        loader = PyPDFLoader(tmp_path)
        docs = loader.load()
        documents.extend(docs)
        os.remove(tmp_path)
    if not documents:
        return jsonify({"error": "No valid PDFs or no text extracted"}), 400
    chain, chat_history = create_conversational_chain(api_key, documents, session_id)
    session_data[session_id] = {"chain": chain, "history": chat_history}
    return jsonify({
        "message": "Documents uploaded and indexed successfully",
        "session_id": session_id
    })


@app.route('/upload', methods=['POST'])
def upload_and_index_pdfs():
    api_key = "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"
    if 'files' not in request.files:
        return jsonify({"error": "No files provided"}), 400
    uploaded_files = request.files.getlist("files")
    extracted_texts = ""
    documents = []
    for uploaded_file in uploaded_files:
        if not uploaded_file.filename.endswith('.pdf'):
            return jsonify({"error": f"Invalid file format: {uploaded_file.filename}. Only PDF allowed."}), 400
        with tempfile.NamedTemporaryFile(delete=False, suffix=".pdf") as tmp:
            uploaded_file.save(tmp.name)
            tmp_path = tmp.name
        with open(tmp_path, "rb") as f:
            extracted_text = extract_text_from_pdf(f.read())
        loader = PyPDFLoader(tmp_path)
        docs = loader.load()
        documents.extend(docs)
        os.remove(tmp_path)
    session_id = str(uuid4())
    chain, chat_history = create_conversational_chain(api_key, documents, session_id)
    session_data[session_id] = {"chain": chain, "history": chat_history}
    save_chat_history(session_id, chat_history)
    return jsonify({
        "message": "PDFs uploaded, extracted, and indexed successfully.",
        "extracted_text": extracted_text,
        "session": session_id
    })


@app.route('/chat', methods=['POST'])
def chat():
    data = request.get_json()
    if not data:
        return jsonify({"error": "No JSON payload provided"}), 400
    session_id = data.get("session_id")
    question = data.get("question")

    if not session_id:
        return jsonify({"error": "Missing session_id"}), 400
    if not question:
        return jsonify({"error": "Missing question"}), 400

    if session_id not in session_data:
        return jsonify({"error": "Session not found. Please upload documents first."}), 404

    chain = session_data[session_id]["chain"]
    chat_history = session_data[session_id]["history"]

    chat_history.add_user_message(question)

    response = chain.invoke({"input": question}, {"configurable": {"session_id": session_id}})

    answer_msg = response.get("answer")
    answer_text = answer_msg.content if hasattr(answer_msg, "content") else str(answer_msg)

    chat_history.add_ai_message(answer_text)
    save_chat_history(session_id, chat_history)

    return jsonify({
        "answer": answer_text,
        "session_id": session_id
    })
```
